```python
x = img.reshape(-1,3)
```

```python
second = x.reshape(img.shape)
```

```python
x.shape
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(10)
```

```python
model.fit(x)
```

```python
color_pellet = model.cluster_centers_
```

```python
for color in color_pellet:
    color_img = np.zeros([100,100,3])
    color_img[:,:] = color
    plt.figure()
    plt.imshow(color_img.astype(int))
    plt.show()
```

```python
gen = np.zeros([x.shape[0],3])
```

```python
for i in range(len(gen)):
    gen[i] = color_pellet[model.labels_[i]]
    
gen = gen.reshape(img.shape)
```

```python
plt.imshow(gen/255)
```
