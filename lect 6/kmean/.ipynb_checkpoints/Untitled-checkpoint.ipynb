```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X , y = make_blobs(1000,centers =6,random_state = 42)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
class KMeans:
    def __init__(self,k=10):
        self.k = k
        self.X = None
        self.clusters = None
    
    def fit(self,X):
        self.X = X
        
        self.clusters = []
        for index in range(self.k):
            cluster = Cluster(X[index])
            self.clusters.append(cluster)
            
        for _ in range(5):
            for point in self.X:
                distances= []
                for cluster in self.clusters:
                    dis = cluster.distance(point)
                    distances.append((dis,cluster))
                distances = sorted(distances)
                chosen = distances[0][1]
                chosen.points.append(point)
            for cluster in self.clusters:
                cluster.update_center()
        
        
class Cluster:
    def __init__(self,center):
        self.center = center
        self.points = []
    
    def update_center(self):
        points = np.array(self.points)
        self.center = np.mean(points,axis=0)
        self.points = []
        
    def distance(self,point):
        return np.sum((self.center - point)**2)
        
```

```python
model = KMeans(4)
```

```python
model.fit(X)
```

```python
plt.scatter(X[:,0],X[:,1])
for cluster in model.cluster:
    plt.scatter(cluster.)
```
