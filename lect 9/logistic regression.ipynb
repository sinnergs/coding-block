```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.linear_model import LogisticRegression
```

```python
X, y = make_blobs(1000, centers=5, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
model = LogisticRegression()
```

```python
model.fit(X, y)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
x1 = np.linspace(-8, 8, 5)
colors = ["red", "blue", "green", "yellow", "orange"]
cs = list(map(lambda x: colors[x], y))
```

```python
plt.figure()
plt.ylim(-10, 10)
plt.scatter(X[:, 0], X[:, 1], c=cs)
i = 0
for coef, inter in zip(model.coef_, model.intercept_):
    slop =  - coef[0] / coef[1]
    intercept = - inter / coef[1]
    x2 = x1 * slop + intercept
    plt.plot(x1, x2, color=colors[i])
    
    i+=1
```

```python
model.score(X, y)
```
