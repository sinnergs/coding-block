```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("taken.jpg")
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
```

```python
fil = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
```

```python
fil
```

```python
res = np.zeros([gray.shape[0]-2,gray.shape[1]-2])
```

```python
gray.shape
```

```python
def apply_filter(gray,fil):
    res = np.zeros([gray.shape[0]-(fil.shape[0]-1),gray.shape[1]-(fil.shape[1]-1)])
    print(res.shape)
    for r in range(res.shape[0]):
        for c in range(res.shape[1]):
            cut = gray[r:r+fil.shape[0],c:c+fil.shape[1]]
            res[r][c]=np.mean(cut*fil)
    return res
```

```python
out = apply_filter(gray,fil)
```

```python
plt.imshow(out,cmap="gray")#edegs
```

```python
fil2 = np.array([[1,1,1],[1,1,1],[1,1,1]])
```

```python
out2 = apply_filter(gray,fil2)
```

```python
plt.imshow(out2)#blur
```

```python
def apply_pool(gray,shape):
    res = np.zeros([gray.shape[0]//shape[0],gray.shape[1]//shape[1]])
    for r in range(0,res.shape[0]):
        for c in range(0,res.shape[1]):
            r_start = r * shape[0]
            c_start = c * shape[1]
            
            cut = gray[r_start:r_start+shape[0],c_start:c_start+shape[1]]
            res[r][c] = np.max(cut)
            
    return res
```

```python
out3 = apply_pool(gray,(2,2)) 
```

```python
plt.imshow(out3)#pixelate
```

```python
out4 = apply_filter(out3,fil)
```

```python
plt.imshow(out4,cmap="gray")
```
