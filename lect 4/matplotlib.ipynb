```python
x = np.linspace(0,2*np.pi,100)
y = np.linspace(-2,2,100)
si = np.sin(x)
co =np.cos(x)
```

```python
plt.figure()
plt.title("A graph btw sin and cos")
plt.plot(x,si,label='sin')
plt.plot(x,co,label= "cos")
plt.xlabel("angles")
plt.ylabel("values")
plt.xlim(0,2*np.pi)
plt.ylim(-2,2)
plt.legend()
plt.show()
```
