```python
def distance(a,b):
    return np.sum((a-b)**2)
```

```python
p1= np.array([1,1])
p2= np.array([4,5])
print(distance(p1,p2))
```

```python
def knn(X,y,point,k):
    
    distances = []
    
    for tup in zip(X,y):#for iterating X and Y simultanously((x1,y1)(x2,y2)........)
        dis = distance(point,tup[0])#tup[0]=X items
        distances.append((dis,tup[1]))#tup[1]= y items
    
    distances = sorted(distances)
    top = np.array(distances[:k])
    print(top)
    vals,occ = np.unique(top[:,1],return_counts = True)
    print(vals,occ)
    #argmax is for index of max item
    
    return(vals[np.argmax(occ)])
    
```
